# Modelo de Predicción para el Precio del Crudo Mexicano

#### Librarías Externas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf
```

#### Rutas e Inputs

```python
inputs = {
    'data_route': './data/',
    'data_filename': 'crudo.xlsx'
}
```

#### Lectura de la Base

Se descarga la base de datos "Precio de la mezcla mexicana de petróleo - Dólares por barril" para los últimos 3 años, en formato XLS, de la página:

https://www.banxico.org.mx/apps/gc/precios-spot-del-petroleo-gra.html

```python
#se lee la base de datos del crudo
data = pd.read_excel(inputs['data_route']+inputs['data_filename'])
```

```python
#se vee un resúmen de la base
data
```

```python
#notamos que hay valores N/E, por lo tanto eliminamos esos renglones
data.drop(data[data['Precio del Petróleo'] == 'N/E'].index, inplace=True)
data = data.reset_index(drop=True)
data.set_index('Fecha', inplace=True)
data
```

#### Gráfica de la Serie de Tiempo

```python
plt.figure(figsize = (10,6))
data['Precio del Petróleo'].plot(figsize=(12, 6))
plt.title('Serie de Tiempo del Precio del Crudo Mexicano, Últimos 3 Años, frecuencia díaria.')
plt.ylabel('Precio en Dólares por Barril')
plt.xlabel('Fecha')
plt.show()
```

#### Análisis de Autocorrelación

```python
# Plot autocorrelation
fig, ax = plt.subplots(figsize=(12, 6))
plot_acf(data['Precio del Petróleo'], lags=30, ax=ax)  # Adjust the number of lags as needed
plt.title('Autocorrelation of Price')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

#### Ajuste de Modelos

Ajuste de Medias Móviles

```python
#medias moviles
data['30-day Moving Average'] = data['Precio del Petróleo'].rolling(window=30).mean()
```

```python
plt.figure(figsize = (10,6))
plt.plot(data.index,data['Precio del Petróleo'],label='Precio Original')
plt.plot(data.index,data['30-day Moving Average'],label='Medias Móviles Mensuales')
plt.title('Serie de Tiempo del Precio del Crudo Mexicano, Últimos 3 Años, frecuencia díaria.')
plt.ylabel('Precio en Dólares por Barril')
plt.xlabel('Fecha')
plt.legend()
plt.show()
```

Predicción del próximo mes usando medias móviles

```python
#extraccion de la última fecha
window_size = 30
last_known_price = data['Precio del Petróleo'].iloc[-1]
last_date = data.index[-1]

#Generar fechas para los próximos 30 días
next_dates = pd.date_range(start=last_date + datetime.timedelta(days=1), periods=30, freq='D')

#crear nuevo data frame para los valores predecidos
predicted_df = pd.DataFrame(index=next_dates, columns=['predicted_price'])

#predecir los valores usando medias móviles
for date in next_dates:
    predicted_price = last_known_price + (date - last_date).days * (data['30-day Moving Average'].iloc[-1] - last_known_price) / window_size
    predicted_df.loc[date, 'predicted_price'] = predicted_price
```

Gráfica de los valores simulados con medias móviles

```python
plt.figure(figsize=(12, 6))
last_100_days = data.index[-100:]  # Get the last 100 days
plt.plot(last_100_days, data['Precio del Petróleo'].loc[last_100_days], label='Precio Original')
plt.plot(predicted_df.index, predicted_df['predicted_price'], label='Valores Simulados')
plt.title('Precios Simulados utilizando Medias Moviles')
plt.ylabel('Precio en Dólares por Barril')
plt.xlabel('Fecha')
plt.legend()
plt.show()
```

Ajuste ARIMA
ahora, realizaremos un ajuste ARIMA de los Datos

Para hacer el ajuste ARIMA, se deben elegir los parámetros (p,d,q) del modelo: autoregressive (p), differencing (d), and moving average (q) 

```python
p = 1
d = 1
q = 1
```

```python
data['Precio del Petróleo'] = pd.to_numeric(data['Precio del Petróleo'], errors='coerce')
data['Precio del Petróleo'].fillna(method='ffill', inplace=True)
```

```python
# Fit the ARIMA model
last_100_days = data.index[-100:]
model = ARIMA(data['Precio del Petróleo'].loc[last_100_days], order=(p, d, q))
model_fit = model.fit()

# Forecast future values
forecast_values = model_fit.forecast(steps=30)  # Predict the next 30 steps

# Create a new dataframe for the predicted values
predicted_ARIMA_df = pd.DataFrame(index=pd.date_range(start=data.index[-1], periods=30, freq='D'), columns=['predicted_price'])
predicted_ARIMA_df['predicted_price'] = forecast_values.values
```

Gráfica de los valores simulados mediante ARIMA

```python
plt.figure(figsize=(12, 6))
last_100_days = data.index[-100:]
plt.plot(last_100_days, data['Precio del Petróleo'].loc[last_100_days], label='Datos Originales')
plt.plot(predicted_ARIMA_df.index, predicted_ARIMA_df['predicted_price'], label='Predicted Values')
plt.title('Datos simulados utilizando ARIMA')
plt.ylabel('Precio en Dólares por Barril')
plt.xlabel('Fecha')
plt.legend()
plt.show()
```

Comparación entre los dos métodos de predicción

```python
plt.figure(figsize=(12, 6))
plt.plot(predicted_df.index, predicted_df['predicted_price'], label='Predicción Medias Moviles')
plt.plot(predicted_ARIMA_df.index, predicted_ARIMA_df['predicted_price'], label='Predicción ARIMA')
plt.title('Datos Simulados: Medias Moviles y ARIMA')
plt.ylabel('Precio en Dólares por Barril')
plt.xlabel('Fecha')
plt.legend()
plt.show()
```
